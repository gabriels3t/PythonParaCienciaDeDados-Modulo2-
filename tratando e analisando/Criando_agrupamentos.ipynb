## Criando Agrupamentos

```python
import pandas as pd
```

```python
dados = pd.read_csv('data/aluguel_residencial.csv', sep=';')
dados.head(10)
```

```python
dados['Valor'].mean()
```

```python
todosOsbairros = dados.Bairro.unique()
todosOsbairros # para pegar todos os bairros 
```

```python
bairros = ['Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca']
selecao = dados['Bairro'].isin(bairros)
dados = dados[selecao]
dados.Bairro.unique()
```

```python
grupo_bairro = dados.groupby('Bairro')
```

```python
type(grupo_bairro)
```

```python
grupo_bairro.groups # é um dicionario com o nome do baixo e o id das casas
```

```python
for bairro, data in grupo_bairro:
    print(f'{bairro} -> {data.Valor.mean()}')
```

```python
grupo_bairro['Valor','Condominio'].mean().round(2)
```

## Estatística Descritivas

```python
grupo_bairro.Valor.describe().round(2) # estatisticas descritivas
```

```python
grupo_bairro.Valor.aggregate(['min','max','sum']).rename(columns={'min':'Minimo','max':'Maximo','sum':'Soma Total'})
```

```python
%matplotlib inline
# configuração a cima pra fazer o matplot funcionar
import matplotlib.pyplot as plt
#configurando
plt.rc('figure',figsize = (20,10))
```

```python
 #std é desfio padrao
grupo_bairro['Valor'].std().plot.bar(color = 'blue')
```
